```python
# import pandas

import pandas as pd
import numpy as np
```

```python
# read csv in
df = pd.read_csv('../data/warn_2019.csv')
df.head(15)
```

```python
df.info()
```

```python
df = df.fillna(0)
```

```python
df = df[df['total_employees'] != 'To'].copy()
```

```python
df = df[df['total_employees'] != 'Currently'].copy()
```

```python
df['affected'] = df['affected'].astype(int)
df['total_employees'] = df['total_employees'].astype(int)
df.info()
```

```python
#reason for closing
df['reason'].unique()
```

```python
df['reason'].value_counts()
```

```python
df.loc[df['reason'].isin(['Plant Closing', 
                          'Plant Unit Closing', 
                          'Possible Plant Closing', 
                          'Plant Closings', 
                          'Possible PlantClosing']), 'reason'] = 'Plant Closing'

df.loc[df['reason'].isin(['Plant Layoff', 
                          'Possible Plant Layoff'
                          ]), 'reason'] = 'Plant Layoff'
```

```python
reasons = df['reason'].value_counts().reset_index().rename(columns = {'index': 'reason', 'reason': 'count'})
reasons
```

```python
reasons.to_csv('../data/reasons_2019.csv', index=False)
```

```python
df.loc[df['union'].isin(['The employees are not represented by a union.',
                         'The employee is not represented by a union.']), 'union'] = 'The employees are not represented by a union' 
df['union'].value_counts().reset_index()
```

```python
#How many people were affected by a WARN notice by business type?
biz_type = df.groupby(['business_type'])['affected'].agg('sum').sort_values(ascending=False).reset_index()
biz_type
```

```python
biz_type.to_csv('../data/biz_type_2019.csv', index=False)
```
